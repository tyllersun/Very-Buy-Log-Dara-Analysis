```python
def create_dataframe(data, ref):
    with open(ref, encoding="utf-8") as f:
        temp = f.read()
        temp = temp.replace("true", '"'+"true"+'"')
        temp = temp.replace("false", '"'+"false"+'"')
        test = temp.split("{")
        #print(len(test))
        for i in range(1, len(test)):
            try:
                dict_ = eval("{" + test[i])
            except:
                continue
#             data.append(dict_, ignore_index=True)
            data.append(dict_)
        return data
```
